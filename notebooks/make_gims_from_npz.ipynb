````python
import matplotlib.pyplot as plt
from ionex_formatter.reader import get_npz_data
from ionex_formatter.ionex_map import GridCell, IonexMap, SpatialRange
from ionex_formatter.formatter import IonexFile, IonexMapType
from pathlib import Path
from datetime import datetime, timedelta
````

````python
path_to_npz = ""
maps, grids = get_npz_data(Path("../tests/data_samples/one_day_maps.npz"))
print(f"Maps have following shape {maps.shape}")
print(f"Grid have keys {list(grids.keys())}")
````

````python
#plt.imshow(maps[16])
````

````python
lon_grid = grids["spatial"][0]
lat_grid = grids["spatial"][1]
maps_cells = list()
for iepoch, epoch in enumerate(grids["time"]):
    vals = maps[iepoch].flatten()
    lats = lat_grid.flatten()
    lons = lon_grid.flatten()
    cells = list()
    for val, lat, lon in zip(vals, lats, lons):
        cell = GridCell()
        cell.lat = float(lat)
        cell.lon = float(lon)
        cell.val = int(round(float(val), 3) * 10)
        cells.append(cell)  
        if lon == -180:
            cell = GridCell()
            cell.lat = float(lat)
            cell.lon = -float(lon)
            cell.val = int(round(float(val), 3) * 10)
            cells.append(cell)  
    maps_cells.append(cells)
            
````

````python
prepared_data = {}
for iepoch, epoch in enumerate(grids["time"]):
    ionex_map = IonexMap(
        lat_range=SpatialRange(87.5, -87.5, -2.5),
        lon_range=SpatialRange(-180, 180, 5),
        height=450,
        epoch=grids["time"][iepoch]
    ) 
    ionex_map.set_data(maps_cells[iepoch])
    prepared_data[epoch] = ionex_map
formatter = IonexFile()
formatter.set_maps(
    prepared_data, 
    dtype=IonexMapType.TEC
)
data_lines = []
for epoch in grids["time"]:
    one_map_lines = formatter.get_map_lines(
        IonexMapType.TEC, 
        epoch
    )
    data_lines.extend(one_map_lines)
````
